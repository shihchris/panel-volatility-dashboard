```python
# =============================================================================
# Robust AR-EWMA Volatility Forecasting
# =============================================================================
import warnings, numpy as np, pandas as pd, matplotlib.pyplot as plt
warnings.filterwarnings("ignore")

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error,
    median_absolute_error, r2_score,
    mean_absolute_percentage_error
)
from sklearn.model_selection import TimeSeriesSplit, ParameterGrid

# ───────── USER CONFIG ───────────────────────────────────────
FEATURE_COLS  = ["abs_volatility","spread","imbalance",
                 "depth","depth_ratio","rv_rolling","mom_rolling"]
LAG_STEP      = 1                 # one 5-min block lag
TARGET_COL    = "realized_volatility"
TIME_COL      = "time_seconds"
DAY_COL       = "day_of_week"     # 1=Mon … 5=Fri
CHUNK_SEC     = 300               # 5-min block
RETRAIN_SEC   = 1800              # 30-min retrain
SCALE_TARGET  = 10_000
N_RANDOM      = 10
RANDOM_SEED   = 42

# AR-EWMA specific parameters
AR_MAX_ORDER = 3                  # Maximum AR order to consider
EWMA_ALPHAS = [0.05, 0.1, 0.2, 0.3, 0.4]  # EWMA decay factors to try

# ───────── LOAD & PREP DATA ─────────────────────────────────
train_df = df_30s.copy()
test_df  = df_30s_test.copy()
full_df  = pd.concat([train_df, test_df]) \
             .sort_values(["stock_id","week",TIME_COL]) \
             .reset_index(drop=True)

# 1) lag features
def lag_features(df, groups, cols, lag=1):
    return df.assign(**{
        f"{c}_lag": df.groupby(groups)[c].shift(lag)
        for c in cols
    })

full_df = lag_features(full_df, ["stock_id","week"], FEATURE_COLS, lag=LAG_STEP)
full_df.dropna(subset=[f"{c}_lag" for c in FEATURE_COLS], inplace=True)
LAG_FEATURES = [f"{c}_lag" for c in FEATURE_COLS]

# 2) scale
scaler = StandardScaler().fit(full_df.loc[full_df[DAY_COL]<5, LAG_FEATURES])
full_df[LAG_FEATURES] = scaler.transform(full_df[LAG_FEATURES])

# ───────── GRID SEARCH TUNER ───────────────────────────────
def tune_ar_ewma(y_train):
    """Find optimal AR order and EWMA alpha parameters"""
    tscv = TimeSeriesSplit(n_splits=3)
    
    # Parameter grid for AR and EWMA
    param_grid = {
        'ar_order': range(1, AR_MAX_ORDER + 1),
        'ewma_alpha': EWMA_ALPHAS,
        'ar_weight': [0.3, 0.4, 0.5, 0.6, 0.7]  # Weight given to AR vs EWMA
    }
    
    best_score = float('-inf')
    best_params = None
    
    # Convert to numpy for faster operations
    y_train_np = y_train.values
    
    for params in ParameterGrid(param_grid):
        scores = []
        
        for train_idx, test_idx in tscv.split(y_train_np):
            y_fold_train, y_fold_test = y_train_np[train_idx], y_train_np[test_idx]
            
            # Fit AR model
            try:
                ar_model = ARIMA(y_fold_train, order=(params['ar_order'], 0, 0))
                ar_fit = ar_model.fit()
                ar_preds = ar_fit.forecast(steps=len(y_fold_test))
                
                # Compute EWMA predictions
                ewma_preds = np.zeros_like(y_fold_test)
                ewma_val = y_fold_train[-1]  # Start with last training value
                
                for i in range(len(y_fold_test)):
                    ewma_preds[i] = ewma_val
                    if i < len(y_fold_test) - 1:
                        # Update EWMA with actual value (known in CV loop)
                        ewma_val = params['ewma_alpha'] * y_fold_test[i] + (1 - params['ewma_alpha']) * ewma_val
                
                # Combine predictions
                combined_preds = params['ar_weight'] * ar_preds + (1 - params['ar_weight']) * ewma_preds
                
                # Compute score
                score = -mean_squared_error(y_fold_test, combined_preds, squared=False)
                scores.append(score)
            except:
                scores.append(float('-inf'))
        
        avg_score = np.mean(scores)
        if avg_score > best_score:
            best_score = avg_score
            best_params = params
    
    print(f"Best AR-EWMA params: {best_params}")
    return best_params

# ───────── METRICS FUNCTION ─────────────────────────────────
def compute_vol_metrics(actuals, preds):
    rmse  = mean_squared_error(actuals, preds, squared=False)
    mae   = mean_absolute_error(actuals, preds)
    medae = median_absolute_error(actuals, preds)
    r2    = r2_score(actuals, preds)
    mape  = mean_absolute_percentage_error(actuals, preds) * 100
    smape = np.mean(
        2 * np.abs(preds - actuals) /
        (np.abs(preds) + np.abs(actuals) + 1e-12)
    ) * 100
    qlike = np.mean(np.log(preds**2) + (actuals**2) / (preds**2))
    return {
        "RMSE": rmse / SCALE_TARGET,
        "MAE": mae / SCALE_TARGET,
        "MedAE": medae / SCALE_TARGET,
        "MAPE(%)": mape,
        "SMAPE(%)": smape,
        "QLIKE": qlike,
        "R2": r2
    }

# ───────── 4. ROLLING FORECAST ───────────────────────────────
def rolling_forecast_ar_ewma(y_base, y_fri, params, step_rows):
    """Rolling forecast using AR-EWMA model"""
    # Initial fit of AR model
    ar_model = ARIMA(y_base, order=(params['ar_order'], 0, 0))
    ar_fit = ar_model.fit()
    
    # Initialize EWMA with last training value
    ewma_val = y_base.iloc[-1]
    
    preds_ar_ewma, preds_naive, actuals, idxs = [], [], [], []
    last_rv = y_base.iloc[-1]  # For naive forecast
    y_hist = y_base.copy()
    
    for st in range(0, len(y_fri), step_rows):
        ed = min(st+step_rows, len(y_fri))
        y_blk = y_fri.iloc[st:ed]
        
        # Generate forecasts for this block
        ar_forecasts = ar_fit.forecast(steps=len(y_blk))
        ewma_forecasts = np.zeros(len(y_blk))
        
        # First EWMA forecast uses current ewma_val
        ewma_forecasts[0] = ewma_val
        
        # Calculate remaining EWMA forecasts (each one using previous forecast)
        for i in range(1, len(y_blk)):
            ewma_forecasts[i] = ewma_forecasts[i-1]  # No actual data for updating
        
        # Combine AR and EWMA forecasts
        combined_forecasts = params['ar_weight'] * ar_forecasts + (1 - params['ar_weight']) * ewma_forecasts
        
        # Store predictions
        preds_ar_ewma.extend(combined_forecasts)
        preds_naive.extend([last_rv] * len(y_blk))
        actuals.extend(y_blk)
        idxs.extend(y_blk.index)
        
        # Update for next iteration
        last_rv = y_blk.iloc[-1]  # Update naive forecast
        y_hist = pd.concat([y_hist, y_blk])
        
        # Update EWMA with the last actual value
        ewma_val = params['ewma_alpha'] * y_blk.iloc[-1] + (1 - params['ewma_alpha']) * ewma_val
        
        # Refit AR model if there are more blocks
        if ed < len(y_fri):
            ar_model = ARIMA(y_hist, order=(params['ar_order'], 0, 0))
            ar_fit = ar_model.fit()
    
    return np.array(preds_ar_ewma), np.array(preds_naive), np.array(actuals), idxs

# ───────── 5. MAIN LOOP ─────────────────────────────────────
metrics = []
first_plot = True
step_rows = int(RETRAIN_SEC / (full_df[TIME_COL].diff().median() or CHUNK_SEC))

for (sid, wk), grp in full_df.groupby(["stock_id","week"]):
    mon_thu = grp[grp[DAY_COL] < 5]
    fri     = grp[grp[DAY_COL] == 5]
    if fri.empty or mon_thu.empty:
        continue

    # For AR-EWMA we only need the target variable
    y_train = mon_thu[TARGET_COL] * SCALE_TARGET
    y_fri   = fri[TARGET_COL] * SCALE_TARGET

    # Tune AR-EWMA parameters
    best_params = tune_ar_ewma(y_train)
    
    # Generate predictions
    pred_ar_ewma, pred_naive, act, idx = rolling_forecast_ar_ewma(
        y_train, y_fri, best_params, step_rows)

    # Compute metrics
    m_a = compute_vol_metrics(act, pred_ar_ewma)
    m_n = compute_vol_metrics(act, pred_naive)

    # Record both
    m_a.update({"stock": sid, "week": wk, "model": "AR-EWMA"})
    m_n.update({"stock": sid, "week": wk, "model": "Naive"})
    metrics.append(m_a)
    metrics.append(m_n)

    # Plot first group
    if first_plot:
        hod = (fri.loc[idx, TIME_COL] % 86400)/3600
        plt.figure(figsize=(12,4))
        plt.plot(hod, act/SCALE_TARGET, label="Actual")
        plt.plot(hod, pred_ar_ewma/SCALE_TARGET, '--', label="AR-EWMA")
        plt.plot(hod, pred_naive/SCALE_TARGET, ':', label="Naive")
        plt.vlines(range(11,18), *plt.ylim(), ls='--', alpha=.3)
        plt.title(f"Stock {sid} Week {wk}")
        plt.xlabel("Hour"); plt.ylabel("Realized Volatility")
        plt.legend(); plt.tight_layout(); plt.show()
        first_plot = False

    print(f"[{sid}-{wk}] AR-EWMA R²={m_a['R2']:.3f} vs Naive R²={m_n['R2']:.3f}")

# ───────── 6. RESULTS TABLE ─────────────────────────────────
met_df = pd.DataFrame(metrics)
fmt = {
    'R2': "{:.3f}".format,
    'RMSE': "{:.2e}".format,
    'MAE': "{:.2e}".format,
    'MedAE': "{:.2e}".format,
    'MAPE(%)': "{:.2f}%".format,
    'SMAPE(%)': "{:.2f}%".format,
    'QLIKE': "{:.3f}".format
}
print("\n=== Metrics (Mon–Thu → Fri, 30-min expanding) ===")
print(met_df.to_string(index=False, formatters=fmt))
```
